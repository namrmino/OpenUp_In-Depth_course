## Groupby X 병렬처리

```python
from multiprocessing import cpu_count , Pool
```

```python
def applyPrallel(dfGrouped , func):
    
    with  Pool(cpu_count()) as p:
        
        ret_list = p.map(func , [group for name , group in dfGrouped])
    return pd.concat(ret_list)
```

```python
cpu_count()
```
